```python
#Task 2: Count Even and Odd Numbers 
a=[10, 15, 22, 35, 40, 51]
evenNumbersCount=0
OddNumbersCount=0
for i in range(len(a)):
    if a[i]%2==0:
        evenNumbersCount+=1
    else:
        OddNumbersCount+=1
print("Even Numbers=",evenNumbersCount)
print("Odd Numbers=", OddNumbersCount)
```

```python
#Task 4: Remove Duplicate Elements 

numbers = [10, 20, 10, 30, 20, 40]
uniqueNumbers=list(set(numbers))
print(uniqueNumbers)
print(type(uniqueNumbers))
```

```python
#Task 5: Separate Positive and Negative Numbers 

numbers = [10, -5, 20, -15, 30, -8]
positiveNumbers=[]
negativeNumbers=[]
for i in range(len(numbers)):
    if numbers[i]>0:
        positiveNumbers.append(numbers[i])
    elif numbers[i]<0:
        negativeNumbers.append(numbers[i])
print("Positive Numbers:", positiveNumbers)
print("Negative Numbers:", negativeNumbers)
```

```python
#Task 6: Reverse a List Without Using reverse()

numbers = [10, 20, 30, 40, 50] 
i=len(numbers)-1
while i>=0:
    print(numbers[i])
    i-=1
```
